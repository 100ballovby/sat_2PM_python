# Типы языков программирования

<p style="font-size: 18px;">Языки делятся на две большие группы:</p>

## Компилируемые языки

<p style="font-size: 18px;">Компилируемый язык программирования позволяет вам написать программный код, который будет трансформирован в машинный код и затем будет иметь возможность запуска на той платформе, для которой его писали.</p>

<img src="screen_1.png">

<p style="font-size: 18px;">Основная мысль заключается в том, что программы, которые написаны с помощью комплилируемых языков программирования, работают жутко быстро (потому что они способны давать инструкции компьютеру на том языке, который он понимает. То есть они просто общаются между собой. Но при этом, чтобы дать программе какую-то новую фишку, ее нужно заново пересобрать, потому что внести изменения в уже скомпилированный код нельзя.</p>

## Интерпретируемые языки

<p style="font-size: 18px;">Интерпретируемые языки программирования характерны тем, что они не разговаривают с компьютером на одном языке. Они имеют возможность запуска после написания первой строчки инструкций. Это обусловлено тем, что в момент запуска программы, которая была написана интерпретируемым языком программирования, <b>интерпретатор</b> (на рисунке синий чел) построчно переводит ваши инструкции в машинный код. Точно так же, как это делает синхронный переводчик. Естественно, что этот процесс занимает определенное время и программы, написанные такими языками, работают немного медленнее, чем скомпилированные.</p>

# Информация и данные

<p style="font-size: 18px;"><b>Информация</b> - сведения об объектах окружающего мира. Эти сведения вы можете воспринимать.</p>
<p style="font-size: 18px;"><b>Данные</b> - информация, которая представлена в закодированном виде. То есть в том виде, в котором ее можно сохранить, обработать и передать.</p>

<p style="font-size: 18px;">Так как способов кодирования информации существует очень много, данным были присвоены типы. <span style="background: yellow;"><b>Тип данных</b> - класс данных, характеризуемый членами класса и операциями, которые могут быть к ним применены.</span></p>

<img src="screen_3.png">

<p style="font-size: 18px;">Важно помнить, что в любом типе данных есть как конечное количество членов, так и конечное количество операций.</p>

<hr>


## Типы данных в Python

<ol style="font-size: 18px;" start="0">
    <li><code>None</code> - <b>пустота</b>. Ничто. Отсутствие. Пустота содержит ровно один элемент - это она сама и операция-вопрос: <i>пусто ли здесь?</i>;</li>
    <li><b>Числа</b>
    <ul>
        <li><code>int</code> - <b>integer</b> - целое число;</li>
        <li><code>float</code> - <b>floating point number</b> - число с десятичной <b><u>точкой</u></b>;</li>
        <li><code>complex</code> - комплексные числа вида <b>1j</b>.</li>
    </ul>
    </li>
    <li><code>bool</code> - <b>boolean</b> - логический тип. Хранит в себе два значения: True/False, используется для проверок;</li>
    <li><code>str</code> - <b>string</b> - строчка. Строкой в программировании называется все, что заключено в кавычки. Для Python не важен тип кавычек, можно использовать как одинарные, так и двойные.</li>
</ol>

<p style="font-size: 18px;">К чему из вышеперечисленного можно отнести <code>+375 (29) 123-45-67</code> и <code>03.12.2008</code>?</p>

<p style="font-size: 18px;">С номером телефона достаточно легко, потому что он является строкой, несмотря на то, что в нем присутствуют числа. С датой все не настолько очевидно, потому что в отличие от многих вещей, дату можно посчитать. В Python присутствует модуль <code>datetime</code> для работы с датами и он как бы добавляет тип данных для даты. В остальных ситуациях дата будет являться строкой (<code>str</code>).</p>
