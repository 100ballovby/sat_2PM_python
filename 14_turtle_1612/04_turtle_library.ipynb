# Библиотека <code>turtle</code>


<p style="font-size: 18px;">Стандартная библиотека Python содержит модуль <code>turtle</code>, предназначенный для обучения базовым конструкциям языка через призму рисования. Модуль содержит набор функций, которые позволяют управлять черепахой на экране и рисовать.</p>

<table style="font-size: 18px;">
    <tr>
        <th>Функция</th>
        <th>Описание</th>
    </tr>
    <tr>
        <td><code>.shape()</code></td>
        <td>Позволяет установить форму черепашки. Доступные: <code>arrow</code>, <code>circle</code>, <code>turtle</code>.</td>
    </tr>
    <tr>
        <td><code>.color('цвет')</code></td>
        <td>Функция позволяет установить цвет, которым рисует черепашка. Цвет может быть прописан как цветовая константа, либо в формате hex-цветов (<code>#3a1f2b</code>). Список цветовых констант <a href="https://cs111.wellesley.edu/labs/lab02/colors" target="_blank">здесь</a>. Есть более приятный вариант - это собрать цвет самостоятельно <a href="https://www.google.com/search?q=color+picker&oq=color+picker&gs_lcrp=EgZjaHJvbWUqDAgAEAAYQxiABBiKBTIMCAAQABhDGIAEGIoFMgcIARAAGIAEMgcIAhAAGIAEMgcIAxAAGIAEMgcIBBAAGIAEMgcIBRAAGIAEMgcIBhAAGIAEMgcIBxAAGIAEMgcICBAAGIAEMgcICRAAGIAE0gEINDIyMWowajeoAgCwAgA&sourceid=chrome&ie=UTF-8" target="_blank">в палитре</a>.</td>
    </tr>
    <tr>
        <td><code>.fillcolor('цвет')</code></td>
        <td>Черепашка умеет заливать нарисованные фигуры цветом. Залить фигуру можно ТОЛЬКО в том случае, если фигура замкнутая (квадрат, круг, труегольник). Все, что было написано относитель функции <code>.color()</code> работает точно так же и здесь.</td>
    </tr>
    <tr>
        <td><code>.begin_fill()</code></td>
        <td>Функция пишется перед циклом рисования замкнутой фигуры для того, чтобы залить эту фигуру цветом, который берется из <code>.fillcolor()</code>. Используется ТОЛЬКО вместе со следующей функцией, отдельное использование не приводит ни к чему.</td>
    </tr>
    <tr>
        <td><code>.end_fill()</code></td>
        <td>Функция пишется после цикла рисования замкнутой фигуры для того, чтобы закончить заливку этой фигуры. Отдельное использование функции не приводит ни к чему.</td>
    </tr>
    <tr>
        <td><code>.fd(x)</code>/<code>.forward(x)</code></td>
        <td>Перемещает черепашку на <code>x</code> "шагов" вперед. <code>x</code> - это положительное целое число, строки не принимаются. Несмотря на то, что функций две, они друг от друга ничем не отличаются. Первая функция - это просто сокращенное название второй.</td>
    </tr>
    <tr>
        <td><code>.bk(x)</code>/<code>.backward(x)</code></td>
        <td>Перемещает черепашку на <code>x</code> "шагов" назад. <code>x</code> - это положительное целое число, строки не принимаются. Несмотря на то, что функций две, они друг от друга ничем не отличаются. Первая функция - это просто сокращенное название второй. Если написать функции <code>.fd()</code> отрицательное значение, черепашка тоже будет ходить назад.</td>
    </tr>
    <tr>
        <td><code>.lt(deg)</code>/<code>.left(deg)</code></td>
        <td>Поворачивает черепашку налево на количество градусов <code>deg</code>. Черепашка поворачивается ТОЛЬКО на месте. Вы не можете идти и одновременно поворачивать.</td>
    </tr>
    <tr>
        <td><code>.rt(deg)</code>/<code>.right(deg)</code></td>
        <td>Поворачивает черепашку направо на количество градусов <code>deg</code>. Черепашка поворачивается ТОЛЬКО на месте. Вы не можете идти и одновременно поворачивать.</td>
    </tr>
    <tr>
        <td><code>.goto(x, y)</code></td>
        <td>Перемещает черепашку в координаты <code>x</code> и <code>y</code>. Координатная плоскость в черепашке классическая. То есть это оси абсцисс и ординат, как в математике. Размер поля (экрана), на котором перемещается черепашка составляет 1000 на 1000 пикселей. Изначально черепашка появляется в точке <code>[0;0]</code>. Движение влево - отрицательный <code>x</code>, движение вправо - положительный <code>x</code>. Движение вверх - положительный <code>y</code>, вниз - отрицательный <code>y</code>. Доступные значения для <code>x</code> и <code>y</code> находятся на промежутках <code>[-500; 500]</code> в обе стороны.</td>
    </tr>
    <tr>
        <td><code>.up()</code>/<code>.penup()</code></td>
        <td>"Поднимает" перо черепашки - черепашка после вызова этой функции перестает рисовать. Используется в связке с функцией <code>.goto(x, y)</code>.</td>
    </tr>
    <tr>
        <td><code>.down()</code>/<code>.pendown()</code></td>
        <td>"Опускает" перо черепашки - черепашка после вызова этой функции начинает рисовать. Используется в связке с функцией <code>.goto(x, y)</code>. <b>По умолчанию перо всегда опущено</b>.</td>
    </tr>
    <tr>
        <td><code>.pensize(width)</code></td>
        <td>Функция устанавливает толщину линии (в пикселях), которой рисует черепашка. <code>width</code> - это положительное челое число.</td>
    </tr>
    <tr>
        <td><code>.speed(x)</code></td>
        <td>Устанавливает скорость перемещения черепашки по полю. Целесообразно использовать только в тех случаях, когда вам нужно рисовать очень много объектов. Принимает в качестве аргумента целые числа от <code>0</code> до <code>9</code>. <code>0</code> - это самая быстрая скорость перемещения.</td>
    </tr>
    <tr>
        <td><code>.stamp()</code></td>
        <td>Функция позволяет оставить на экране "отпечаток" формы черепашки.</td>
    </tr>
    <tr>
        <td><code>.circle(rad, extent)</code></td>
        <td>Рисует <b>окружность</b> с радиусом <code>rad</code>. Окружность всегда рисуется из верхней точки. Если передать параметр <code>extent</code>, можно нарисовать полу-/четверть-/и тд-окружности. Например, вызов <code>.circle(100, 180)</code>, нарисует полукруг с радиусом 100.</td>
    </tr>
    <tr>
        <td><code>.dot(dia)</code></td>
        <td>Рисует <b>круг</b> или точку, с диаметром <code>dia</code>. Точка имеет тот же цвет, который имеет и сама черепашка.</td>
    </tr>
    
</table>

# Рисование простых фигур

<p style="font-size: 18px;">За базовые фигуры берем квадрат, прямоугольник и равнобедренный треугольник. Рисование квадрата и прямоугольника можно оформить большим количеством перемещений черепашки вперед и поворотами, но гораздо логичнее в этом случае пользоваться циклами.</p>
